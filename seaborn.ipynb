```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
print(tips.head(n=20))
```

```python
menSuntip = tips.loc[(tips['sex']=='Male')&(tips['day']=='Sun'),'tip']
womenSuntip = tips.loc[(tips['sex']=='Female')&(tips['day']=='Sun'),'tip']
menSattip = tips.loc[(tips['sex']=='Male')&(tips['day']=='Sat'),'tip']
womenSattip = tips.loc[(tips['sex']=='Female')&(tips['day']=='Sat'),'tip']
```

```python
print(menSuntip)
```

```python
plt.hist(menSuntip, color='red',  alpha=.5,label='Men on Sun')  # alpha just controls the opacity
plt.hist(womenSuntip, color='blue', alpha=.5,label='Women on Sun')
plt.hist(menSattip, color='green', alpha=.5,label='Men on Sat')  
plt.hist(womenSattip, color='yellow', alpha=.5,label='Women on Sat')
plt.xlabel('Tip in USD')
plt.legend(loc='upper right')
plt.title('Plot 1: Meal Tips by Gender and day')
plt.show()
```

```python
sns.set(style="ticks")
```

```python
g = sns.FacetGrid(tips, row="sex", col="day")
g.map(plt.hist, "tip", color="red",  lw=0)
plt.subplots_adjust(top=0.9)
g.set_axis_labels('Tip in USD')
plt.suptitle('Plot 2: Meal Tips by Gender and day')
sns.despine(trim=True)
plt.show()
```

```python
ax = sns.boxplot(x='day',y='tip',hue='sex',data=tips)  
plt.title('Plot 3: Meal Tips by Gender and Time')
sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='Tips')

plt.show()
```

```python
sns.set(style="darkgrid")

g = sns.barplot(x="sex", y="tip", hue="day", data=tips,
                palette="pastel", ci=95)

plt.title('Barplot: Meal Tips by day and Gender')
plt.show()
```

```python
sns.boxenplot(x="sex", y="tip", hue = 'day',
              color="b",
              scale="linear", data=tips)
```

```python
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
day_order = ["Thur", "Fri", "Sat", "Sun"]
sns.scatterplot(x="total_bill", y="tip",
                hue="day", 
                palette="ch:r=-.2,d=.3_r",
                hue_order=day_order,
                sizes=(1, 8), linewidth=0,
                data=tips, ax=ax)
```

```python
sns.lineplot(x="total_bill", y="tip",
             hue="day", 
             data=tips)
```

```python
sns.boxenplot(x="total_bill", y="tip", hue = 'day',
              color="b", data=tips)
```
